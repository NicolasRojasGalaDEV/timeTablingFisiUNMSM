```python
import pandas as pd
```

```python
df=pd.read_csv('..\\Data\\Extraidos\\sum_horario_programado2025I.csv',sep=';',encoding='utf-8-sig')
df.head()   
```

```python
print(df.dtypes)
```

```python
df["HoraInicioFormateada"] = pd.to_datetime(df["HoraInicio"], format="%H:%M")
#pd.to_datetime(df["HoraInicioFormateada"], format="%H:%M")
df
```

```python
df["HoraFinFormateada"] = pd.to_datetime(df["HoraFin"], format="%H:%M")
df
```

```python
print(df.dtypes)
```

```python
df["Duracion"] = df["HoraFinFormateada"] - df["HoraInicioFormateada"]
df
```

```python
df["DuracionHoras"] = df["Duracion"].dt.total_seconds() / 3600
df
```

```python
horas_dictadas_docente=df.groupby(["CodAsignatura","NomAsignatura", "CodDocente","NomDocente", "Sección"])["DuracionHoras"].sum().reset_index()
horas_dictadas_docente
```

## Se genera el dataset Duracion Asignatura

```python
duracion_asignatura = horas_dictadas_docente[["CodAsignatura","NomAsignatura", "DuracionHoras"]].drop_duplicates()
duracion_asignatura
```

Agregamos dato de a que curso pertenece la asignatura

```python
ciclo_asignatura=pd.read_csv("..\\Data\\Extraidos\\ciclo_asignatura.csv",sep=";",encoding="utf-8-sig")
ciclo_asignatura.head()
```

```python
duracion_asignatura = duracion_asignatura.merge(
    ciclo_asignatura[['CodAsignatura', 'Ciclo']],
    how='left',
    on='CodAsignatura'
).drop(columns='CodAsignatura')  # elimina el campo CodAsignatura si no lo necesitas
```

```python
duracion_asignatura.head()
```

```python
duracion_asignatura= duracion_asignatura.merge(
    ciclo_asignatura[['NomAsignatura','CodAsignatura','Ciclo']],
    how='left',
    on='NomAsignatura'
)
```

```python
duracion_asignatura
```

```python
duracion_asignatura = duracion_asignatura.drop(columns=['Ciclo_y'])  # eliminar la columna Ciclo_y
duracion_asignatura = duracion_asignatura.rename(columns={'Ciclo_x': 'Ciclo'})  # renombrar Ciclo_x como Ciclo
duracion_asignatura.head()
```

```python
duracion_asignatura.to_csv("..\\Data\\Extraidos\\duracion_asignatura.csv", index=False,sep=";",encoding='utf-8-sig')
```
