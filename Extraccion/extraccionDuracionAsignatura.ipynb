```python
import pandas as pd
```

```python
df=pd.read_csv('..\\Data\\Extraidos\\sum_horario_programado2025I.csv',sep=';',encoding='utf-8-sig')
df.head()   
```

```python
print(df.dtypes)
```

```python
df["HoraInicioFormateada"] = pd.to_datetime(df["HoraInicio"], format="%H:%M")
#pd.to_datetime(df["HoraInicioFormateada"], format="%H:%M")
df
```

```python
df["HoraFinFormateada"] = pd.to_datetime(df["HoraFin"], format="%H:%M")
df
```

```python
print(df.dtypes)
```

```python
df["Duracion"] = df["HoraFinFormateada"] - df["HoraInicioFormateada"]
df
```

```python
df["DuracionHoras"] = df["Duracion"].dt.total_seconds() / 3600
df
```

```python
#df[
    #(df["Asignatura"] == "202W0507 - INGENIERÍA DE REQUISITOS") 
    #& (df["Docente"] == "0A1609 - ZAVALETA CAMPOS, JORGE LUIS")
#]
```

```python
horas_dictadas_docente=df.groupby(["CodAsignatura","NomAsignatura", "CodDocente","NomDocente", "Sección"])["DuracionHoras"].sum().reset_index()
horas_dictadas_docente
```

## Se genera el dataset Duracion Asignatura

```python
duracion_asignatura = horas_dictadas_docente[["CodAsignatura","NomAsignatura", "DuracionHoras"]].drop_duplicates()
duracion_asignatura
```

```python
duracion_asignatura.to_csv("..\\Data\\Extraidos\\duracion_asignatura.csv", index=False,sep=";",encoding='utf-8-sig')
```
