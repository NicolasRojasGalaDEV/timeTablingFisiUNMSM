```python
import pandas as pd
```

```python
df=pd.read_csv("..\\Data\\Extraidos\\sum_horario_programado2025I.csv",sep=";",encoding='utf-8-sig')
df.head()
```

```python
df=df.drop(columns=["CodDocente","NomDocente","Sección","Aula","Día","HoraInicio","HoraFin","Tope"])
df.head()
```

```python
volumen_matriculados=df.groupby(["CodAsignatura","NomAsignatura","Créditos"])["Matriculados"].sum().reset_index()
volumen_matriculados
```

```python
df=df[["CodAsignatura","NomAsignatura","Créditos"]].drop_duplicates()
df
```

```python
df = pd.merge(df, volumen_matriculados, on=["CodAsignatura", "NomAsignatura", "Créditos"], how="left")
df
```

```python
df.to_csv("..\\Data\\Extraidos\\volumen_estudiantes_actual.csv",index=False,sep=";",encoding='utf-8-sig')
```
