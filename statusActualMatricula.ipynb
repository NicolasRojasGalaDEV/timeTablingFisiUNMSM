```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Data\\matriculados2025I.csv', sep=';')
df.head()
```

```python
df.columns
```

```python
df[['CodigoDocente', 'NombreDocente']] = df['Docente'].str.split(' - ', n=1, expand=True)
df.head()
```

```python
df[['CodigoAsignatura', 'NombreAsignatura']] = df['Asignatura'].str.split(' - ', n=1, expand=True)
df.head()
```

```python
df_filtrado = df[df['Matriculados'] > df['Tope']]
df_filtrado
```

```python
dfFiltrado = df[df['Matriculados'] > 40].sort_values(by='Matriculados', ascending=False)
dfFiltrado
```

```python
cantidadMatriculados=df['Matriculados'].sum()
cantidadMatriculados
```

```python
matriPorAsig = df.groupby('Asignatura')['Matriculados'].sum().reset_index()
matriPorAsig.columns = ['Asignatura', 'cantidadMatriculadosTotal']
matriPorAsig = matriPorAsig.sort_values(by='cantidadMatriculadosTotal', ascending=False)
matriPorAsig    
```

```python
salonPorAsig = df['Asignatura'].value_counts().reset_index()
salonPorAsig.columns = ['Asignatura', 'CantidadAulasAsIs']
salonPorAsig = salonPorAsig.sort_values(by='CantidadAulasAsIs', ascending=False)
salonPorAsig
```

```python
dfMerged = pd.merge(matriPorAsig, salonPorAsig, on='Asignatura', how='left')
dfMerged["CantidadAulasToBe"] = (dfMerged["cantidadMatriculadosTotal"] / 40).round(0).clip(lower=1)
dfMerged
```

```python
difTobeAsIs = dfMerged["CantidadAulasToBe"] > dfMerged["CantidadAulasAsIs"]
dfDiferencias = dfMerged[difTobeAsIs]
dfDiferencias
```

```python
df['mMayor40'] = np.where(df['Matriculados'] > 40, 1, 0)
df['mMenor40'] = np.where(df['Matriculados'] < 40, 1, 0)
df
```

```python
tieneSumMenor40m = df.groupby('Asignatura')['mMenor40'].sum().reset_index()
tieneSumMenor40m.rename(columns={'mMenor40': 'cursosMenor40m'}, inplace=True)
tieneSumMenor40m
```

```python
tieneSumMayor40m = df.groupby('Asignatura')['mMayor40'].sum().reset_index()
tieneSumMayor40m.rename(columns={'mMayor40': 'cursosMayor40m'}, inplace=True)
tieneSumMayor40m[tieneSumMayor40m["cursosMayor40m"]>=1]
```

```python
df = pd.merge(df,dfMerged, on='Asignatura', how='left')
df
```

```python
df = pd.merge(df,tieneSumMenor40m, on='Asignatura', how='left')
dfMerged = pd.merge(df,tieneSumMenor40m, on='Asignatura', how='left')
dfMerged = pd.merge(df,tieneSumMayor40m, on='Asignatura', how='left')
dfMerged
```

```python
flgMalaDistribucion = np.where((dfMerged['cursosMenor40m'] >= 1) & (dfMerged['cursosMayor40m'] >= 1), 1, 0)
dfMerged["flgMalaDistribucion"] = flgMalaDistribucion
dfMerged
#dfMerged[dfMerged["flgMalaDistribucion"]>=1]
```

```python
dfMerged['%concentracionAlumnos']=dfMerged["Matriculados"]/dfMerged["cantidadMatriculadosTotal"]*100
asigMalaDistribucion=dfMerged[(dfMerged["flgMalaDistribucion"] >= 1) & (dfMerged['%concentracionAlumnos'] >= 50)]
asigMalaDistribucion
```

```python
asigMalaDistribucion=dfMerged[(dfMerged["flgMalaDistribucion"] >= 1) & (dfMerged['%concentracionAlumnos'] >= 50)]
listAsigMalaDistribucion=asigMalaDistribucion["Asignatura"].unique()
listAsigMalaDistribucion
```

```python
dfFiltrado = dfMerged[dfMerged['Asignatura'].isin(listAsigMalaDistribucion)]
dfFiltrado
```

```python
import matplotlib.pyplot as plt
```

```python
# Crear columna Docente-Sec
dfFiltrado['DocenteSec'] = dfFiltrado['NombreDocente'].str.strip() + ' - ' + dfFiltrado['Sec.'].astype(str).str.strip()

# Iterar sobre asignaturas
for asignatura in dfFiltrado['NombreAsignatura'].unique():
    df_asig = dfFiltrado[(dfFiltrado['NombreAsignatura'] == asignatura) & (dfFiltrado['CantidadAulasAsIs'] != 1)]
    
    plt.figure(figsize=(6, 8))
    bars = plt.bar(df_asig['DocenteSec'], df_asig['Matriculados'], color='skyblue')
    plt.title(f'{asignatura} - Cantidad de estudiantes por docente-sección')
    plt.xlabel('Docente - Sec')
    plt.ylabel('Cantidad de Matriculados')
    plt.xticks(rotation=45, ha='right')
    
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{int(height)}',
                 ha='center', va='bottom', fontsize=9)
    
    plt.tight_layout()
    plt.show()
```
